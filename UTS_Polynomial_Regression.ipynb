## Preprocessing Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_excel('HARGA RUMAH JAKSEL.xlsx')
```

```python
new_column = [col.replace('HARGA', 'harga').replace('KOTA', 'kota') for col in data.columns]
data.columns = new_column
```

```python
new_data = ['LT', 'LB', 'JKT', 'JKM', 'GRS', 'harga']
data = data[new_data]
data.head()
```

```python
def transform_grs (grs):
  if grs == 'ADA':
    return 1
  if grs == 'TIDAK ADA' :
    return 0
data['GRS'] = data['GRS'].apply(transform_grs)
```

```python
data.head()
```

```python
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
data = pd.DataFrame(data, columns = ['LT', 'LB', 'JKT', 'JKM', 'GRS', 'harga'])
```

```python
data.head()
```

## Polynomial Regression

```python
import math
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
x = data[['LT', 'LB', 'JKT', 'JKM', 'GRS']]
y = data[['harga']]
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
x_poly = poly.fit_transform(x)
model = LinearRegression()
model.fit(x_poly, y)
y_pred = model.predict(x_poly)
```

```python
r_squared = r2_score(y, y_pred)
print("R-squared:", r_squared)
```

```python
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```

```python
rmse = math.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```
